I am here creating a tree of popular tv series namely "Game of Thrones" with specifically "Stark Family".


```python
import graphviz

dot = graphviz.Digraph()
dot.node('A', 'Richard Stark')
dot.node('B', 'Brandon Stark')
dot.node('C', 'Edward Stark') # ed
dot.node('J', 'Katelyn Tully') # kat
dot.node('Z', 'Edward and Katelyn')
dot.node('D', 'Benjen Stark')
dot.node('E', 'Lyanna Stark')
dot.node('F', 'Rhaegar Targaryen')
dot.node('G', 'Robb Stark')
dot.node('H', 'Sansa Stark')
dot.node('I', 'Arya Stark')
dot.node('K', 'Bryan Stark')
dot.node('L', 'Rickon Stark')
dot.node('M', 'Jon Snow')

dot.edges(['AB', 'AC','AD','AE','CZ','JZ','ZG','ZH','ZI','ZK','ZL','EM','FM'])
dot
```
